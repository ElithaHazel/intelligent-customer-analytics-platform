## Project Objective

Olist generates large volumes of transactional 
and customer interaction data but lacks a unified 
analytics platform to identify high-value customers, 
understand purchasing behavior, and support 
proactive business decision-making.

This project builds an end-to-end Customer Lifetime 
Value (CLV) prediction platform on Databricks that:

1. Ingests and processes 100K+ orders across 9 
   relational tables using a Medallion Lakehouse 
   architecture (Bronze, Silver, and Gold layers)

2. Performs feature engineering by combining 
   transactional, behavioral, and RFM (Recency, 
   Frequency, Monetary) features to segment 
   customers into High, Medium, and Low value groups

3. Develops and evaluates an XGBoost-based CLV 
   classification model to predict future 
   high-value customers

4. Tracks all ML experiments and model lifecycle 
   using MLflow on Databricks

5. Provides a RAG assistant that enables business 
   users to interact with customer reviews and 
   business insights using natural language queries

## Dataset — Brazilian E-Commerce (Olist)

**Source:** Kaggle — olistbr/brazilian-ecommerce  
**Size:** ~100,000 real orders | 9 relational tables | 52 columns  
**Period:** 2016–2018  

| Table | Description |
|-------|-------------|
| olist_customers_dataset | Customer info and location |
| olist_orders_dataset | Order status and timestamps |
| olist_order_items_dataset | Products purchased per order |
| olist_order_payments_dataset | Payment method and values |
| olist_order_reviews_dataset | Customer reviews and text ← RAG layer |
| olist_products_dataset | Product catalog and categories |
| olist_sellers_dataset | Seller information and location |
| olist_geolocation_dataset | ZIP code coordinates |
| product_category_name_translation | English category names |

**Why Olist?**
- 9 relational tables = real-world data engineering complexity
- Customer review text = foundation for RAG and semantic search
- Rich transactional data = perfect for ML segmentation and forecasting

## Architecture — Medallion Lakehouse

**Data Flow:**

**RAW** → 9 Olist CSV Files (Kaggle)

**🥉 BRONZE LAYER**
- Raw ingestion with no transformations
- Add metadata columns: ingestion_date, source_file, batch_id
- Save all 9 tables as Delta format

**🥈 SILVER LAYER**
- Remove nulls and duplicates
- Fix data types and standardize formats
- Add derived columns: TotalRevenue, DeliveryDays
- Save as partitioned Delta tables

**🥇 GOLD LAYER**
- Aggregated business metrics
- RFM features for customer segmentation
- Time series data for revenue forecasting
- ML-ready feature store

**🤖 ML MODELS**
- Customer Segmentation (K-Means Clustering)
- Revenue Forecasting (Time Series)
- Churn Prediction (Classification)

**📚 RAG KNOWLEDGE ASSISTANT**
- Customer review embeddings
- Vector semantic search
- LLM-powered Q&A over reviews
